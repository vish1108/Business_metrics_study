```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\vishal\OneDrive\Desktop\case study\WA_Fn-UseC_-Telco-Customer-Churn.csv")

# Display the first few rows of the dataset
df.head(5)
```

```python
# Summary of the dataset
df.info()
```

```python
#chaek the missing value

df.isnull().sum()
```

```python
#Calculate the overall churn rate
churn_rate = df['Churn'].value_counts(normalize=True)['Yes'] * 100
print(f"\nOverall Churn rate {churn_rate:.2f}%")
```

```python
#Plot the churn distribution

plt.figure(figsize=(6, 4))
sns.countplot(data = df, x = "Churn", palette="Set2")
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
# Analyze churn rate by tenure
df['tenure_group'] = pd.cut(df['tenure'], bins = [0, 12, 24, 36, 48, 60, 72],
                           labels=['0-12', '12-24', '24-36', '36-48', '48-60', '60-72'])

churn_by_tenure = df.groupby('tenure_group')['Churn'].value_counts(normalize=True).unstack()['Yes'] * 100

print("\nChurn Rate by Tenure Group :")
print(churn_by_tenure)
```

```python
#plot churn rate by tenure group 
plt.figure(figsize=(8, 5))
churn_by_tenure.plot(kind = 'bar', color = 'skyblue')
plt.title('Churn rate by tenure group')
plt.xlabel('Tenure Group (Months)')
plt.ylabel('Churn rate (%)')
plt.xticks(rotation  = 45)
plt.show()
```

```python
# Analyze churn rate by contract type
churn_by_contract = df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()['Yes'] * 100

print("\nChurn rate by Contract Type: ")
print(churn_by_contract)
```

```python
#plot churn rate by contract type
plt.figure(figsize=(8, 5))
churn_by_contract.plot(kind = 'bar', color = ['orange', 'green', 'red'])
plt.title('Churn Rate by Contract Type')
plt.xlabel("Contract Type")
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation = 0)
plt.show()
```

```python
#Analyze churn rate by payment method
churn_by_payment = df.groupby('PaymentMethod')['Churn'].value_counts(normalize=True).unstack()['Yes'] * 100

print("\nChurn Rate by payment method")
print(churn_by_payment)
```

```python
# Plot churn rate by payment method
plt.figure(figsize=(10, 6))
churn_by_payment.plot(kind = 'bar', color = 'purple')
plt.title("Churn Rate by payment method")
plt.xlabel("Payment Method")
plt.ylabel("Churn Rate (%)")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Summary of finding

print("\nSummary of Insights :")
print(f"- overall Churn rate is {churn_rate: .2f}%")
print("- Customers with shorter tenure tend to have higher churn rates.")
print("- Month-to-month contracts show significantly higher churn rates compared to longer-term contracts.")
print("- Electronic check payment method is associated with the highest churn rate among payment methods.")
```
